````python
# Write a program showing multilevel inheritance.
# 1. Create a base class Vehicle with a method displayInfo().
# 2. Derive a class Car from Vehicle and override displayInfo().
# 3. Further derive a class ElectricCar from Car and again override displayInfo() to show specific details.
class Vehicle:
  def displayInfo(self):
    print("This is a Vehicle")
class Car(Vehicle):
  def displayInfo(self):
    print("This is a Car")
class ElectricCar(Car):
  def displayInfo(self):
    print("This is an Electric Car")
    super().displayInfo()
    Car.displayInfo(self)
````

````python
class bank:
  def __init__(self, name, balance):
   self.name = name
   self._balance = balance
  def deposit(self, amount):
    self._balance += amount
  def withdraw(self, amount):
    if amount <= self._balance:
      self._balance -= amount
    else:
      print("Insufficient balance")
````

````python
from abc import ABC, abstractmethod

# Abstract Class
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
  @abstractmethod
  def perimeter(self):
    pass

# Child Class
class Rectangle(Shape):
  def __init__(self, l, b):
    self.l = l
    self.b = b
  def area(self):
    print("Area is: ", self.l*self.b)
  def perimeter(self):
    print("Perimeter is: ",2*(self.l+self.b))
class Circle(Shape):
  def __init__(self, r):
    self.r = r
  def area(self):
    print("Area is: ", 3.14*self.r*self.r)
  def perimeter(self):
    print("Perimeter is: ",2*3.14*self.r)
````

## Formulate Hypotheses

### Null and Alternative Hypotheses

## Calculate T-Statistic

### Subtask:
Calculate the t-statistic for a one-sample t-test using the provided sample data and population mean.


**Reasoning**:
First, I will set up the necessary variables and import the `numpy` library to prepare for the t-statistic calculation, as per the instructions. Since the values for `population_mean`, `sample_mean`, `sample_size`, and `sample_std` are not explicitly provided in the prompt, I will use placeholder values. If these values were obtained in a previous 'Understand the Data' step, the user should replace the placeholder values with the correct ones.



````python
import numpy as np

# Define the variables with placeholder values (replace with actual values from 'Understand the Data' step)
population_mean = 50  # Example population mean
sample_mean = 52      # Example sample mean
sample_size = 30      # Example sample size
sample_std = 5        # Example sample standard deviation

print("Variables initialized and numpy imported.")
````

**Reasoning**:
Now that the variables are defined and `numpy` is imported, I will calculate the t-statistic using the formula provided in the instructions and print the result.



````python
t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
print(f"The calculated t-statistic is: {t_statistic}")
````

## Determine P-value or Critical Value

### Subtask:
Determine the p-value associated with the calculated t-statistic or the critical t-value for a 95% confidence level and given degrees of freedom.


```markdown
**Reasoning**:
To determine whether the sample average significantly affects the salary, we need to compare our calculated t-statistic with a critical t-value. This involves defining a significance level (alpha), calculating the degrees of freedom, and then using the t-distribution to find the critical value corresponding to our chosen confidence level. We will use `scipy.stats` for this.

### Instructions
1. Define the significance level (alpha) as 0.05. This represents a 95% confidence level.
2. Calculate the degrees of freedom, which is `sample_size - 1`.
3. Import the `t` object from `scipy.stats` to access the t-distribution functions.
4. Calculate the critical t-value for a two-tailed test using the `ppf` (percent point function or quantile function) method of the t-distribution. The arguments will be `1 - alpha/2` and the calculated degrees of freedom.
5. Print the calculated degrees of freedom and the critical t-value.
```

**Reasoning**:
I will now implement the instructions to calculate the degrees of freedom and the critical t-value using `scipy.stats`.



````python
from scipy import stats

# Define the significance level (alpha)
alpha = 0.05

# Calculate degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the critical t-value for a two-tailed test
critical_t_value = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)

print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"Critical t-value (for a 95% confidence level, two-tailed): {critical_t_value}")
````

**Reasoning**:
Now that the critical t-value has been calculated, I will calculate the p-value for the two-tailed test using the t-statistic and degrees of freedom. This will provide an alternative way to determine statistical significance.



````python
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), degrees_of_freedom))
print(f"The calculated p-value is: {p_value}")
````

## Draw Conclusion for Hypothesis Test

### Subtask:
Compare the p-value with the significance level (0.05) or the t-statistic with the critical t-value to conclude whether the sample significantly affects the salary of the people.


**Reasoning**:
To draw a conclusion for the hypothesis test, I will compare the calculated p-value with the significance level and the absolute t-statistic with the critical t-value. This will help determine whether the null hypothesis should be rejected or failed to be rejected, and interpret the findings in the context of the problem.



````python
print("\n--- Conclusion based on p-value ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: The sample average significantly affects the salary of the people.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: The sample average does not significantly affect the salary of the people.")

print("\n--- Conclusion based on t-statistic and critical t-value ---")
if abs(t_statistic) > critical_t_value:
    print(f"Since the absolute t-statistic (|{t_statistic:.4f}|) is greater than the critical t-value ({critical_t_value:.4f}), we reject the null hypothesis.")
    print("Conclusion: The sample average significantly affects the salary of the people.")
else:
    print(f"Since the absolute t-statistic (|{t_statistic:.4f}|) is less than or equal to the critical t-value ({critical_t_value:.4f}), we fail to reject the null hypothesis.")
    print("Conclusion: The sample average does not significantly affect the salary of the people.")
````
